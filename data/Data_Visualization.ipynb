# Data Visualization of the Results

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
participant_stat = pd.read_csv("Participant_Descriptive_Stat.csv")
```

```python
participant_stat.set_index("Unnamed: 0",inplace=True)
```

## 1. Song Periods vs. Recalled Songs

```python
df_song_recall = pd.read_csv("SongPeriods_DependentVar_MEAN.csv")
```

```python
df_song_recall_std = pd.read_csv("SongPeriods_DependentVar_STD.csv")
```

```python
df_song_recall.set_index("Unnamed: 0", inplace=True)
```

```python
df_song_recall
```

```python
x_axis = ["1970-1974",
         "1975-1979",
         "1980-1984",
         "1985-1989",
         "1990-1994",
         "1995-1999",
         "2000-2004",
         "2005-2009"]
```

```python
df_song_recall.index = x_axis
```

### Merge Parent's Means

```python
parents_mean = (participant_stat.loc["mean","Mother Birthdate"] + participant_stat.loc["mean","Father Birthdate"]) /2
```

```python
plt.figure(figsize=(15,8))
plt.plot(df_song_recall.index,df_song_recall["Recalled Songs"])
plt.axvline(participant_stat.loc["mean","Participant Birthdate"], linestyle="dashed",c="m", label="Participant Birthdate")
plt.axvline(parents_mean+20, linestyle="dashed",c="y", label="Parents 20 years old")
plt.ylim([0,4])
plt.xlabel("5 Year Period")
plt.ylabel("Ratings")
#plt.title("Song Periods vs. Recalled Songs")
plt.legend()
plt.show()
```

## 2. Song Periods vs. Listening Type

```python
df_song_recall.columns
```

```python
plt.figure(figsize=(18,8))
plt.plot(df_song_recall.index,df_song_recall["Childhood"], label="Childhood")
#plt.plot(df_song_recall.index,df_song_recall["Recent"], label="Recently")
plt.plot(df_song_recall.index,df_song_recall["Parents"], label="Parents")
plt.plot(df_song_recall.index,df_song_recall["Other People"], label="Other People")
#plt.plot(df_song_recall.index,df_song_recall["Alone"], label="Alone")
plt.xlim([1970,2005])

plt.axvline(participant_stat.loc["mean","Participant Birthdate"], linestyle="dashed",c="m", label="Participant Birthdate")
plt.axvline(parents_mean+20, linestyle="dotted",c="c", label="Parent 20 years old")
plt.ylim([0,4])
plt.xlabel("5 Year Period")
plt.ylabel("Ratings")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(18,8))
plt.plot(df_song_recall.index,df_song_recall["Recent"], label="Recently")
plt.plot(df_song_recall.index,df_song_recall["Alone"], label="Alone")
plt.xlim([1970,2005])

plt.axvline(participant_stat.loc["mean","Participant Birthdate"], linestyle="dashed",c="m", label="Participant Birthdate")
plt.axvline(parents_mean+20, linestyle="dotted",c="c", label="Parent 20 years old")
plt.ylim([0,4])
plt.xlabel("5 Year Period")
plt.ylabel("Ratings")
plt.legend()
plt.show()
```

## In the figure above, we can see that the first two listening type is childhood and parents matches the vertical line where the parent's age is 20.

# 3. Vividness and Evoked Memories

```python
df_song_recall
```

```python
df_song_recall["Evoke Memory"] = df_song_recall["Evoke Memory"]*100
```

```python
df_song_recall["Vividness"] = df_song_recall["Vividness"] *100
```

```python
fig, ax = plt.subplots(figsize=(15,10))
ind = np.arange(8)
ax = fig.add_subplot(111)
ax.bar(ind, df_song_recall["Evoke Memory"], color="c", label="Evoked Memories", width=0.27)
ax.bar(ind+0.3, df_song_recall["Vividness"], color="r", label="Vividness", width=0.27)
ax.plot(ind,df_song_recall["Evoke Memory"], color="r")
ax.set_xticks(ind+0.27)
ax.set_xticklabels(x_axis)

plt.ylabel("% Percentage")
plt.xlabel("Song Periods")
plt.legend()
#plt.title("Evoked Memories & Vividness")
plt.show()
```

```python
plt.bar(df_song_recall.index,df_song_recall["Evoke Memory"],color="b", label="Evoked Memories")
#plt.bar(df_song_recall.index,df_song_recall["Vividness"],color="c", label="Vividness")
plt.plot(df_song_recall.index, df_song_recall["Evoke Memory"], color="r")
plt.ylim([0,100])
```

```python
plt.bar(df_song_recall.index,df_song_recall["Vividness"],color="c", label="Vividness")
```

# For the figure above, you can uncomment the second plt line and see the evoked + vividness all together on the same plot.

# 4. Age Gap

```python
df_age_1 = pd.read_csv("AgeGap_1.csv")
df_age_2 = pd.read_csv("AgeGap_2.csv")
df_age_3 = pd.read_csv("AgeGap_3.csv")
```

```python
df_age_1.set_index("Unnamed: 0", inplace=True)
df_age_2.set_index("Unnamed: 0", inplace=True)
df_age_3.set_index("Unnamed: 0", inplace=True)
```

```python
plt.figure(figsize=(15,5))
plt.plot(df_age_1.index, df_age_1["Recalled Songs"], color="c")
plt.axvline(df_age_1["Parent 1 Birthdate"].mean()+20,label="Parent 1 at his/her 20's",color="c",linestyle="dashed")
plt.axvline(df_age_1["Parent 2 Birthdate"].mean()+20,label="Parent 2 at his/her 20's", color="c",
           linestyle="dashed")
plt.xlabel("Song Periods")
plt.ylabel("Recalled Songs Ratings")
plt.ylim([0,5])
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15,5))

plt.plot(df_age_2.index, df_age_2["Recalled Songs"], color="c")
plt.axvline(df_age_2["Parent 1 Birthdate"].mean()+20,label="Parent 1 at his/her 20's",color="r",linestyle="dashed")
plt.axvline(df_age_2["Parent 2 Birthdate"].mean()+20,label="Parent 2 at his/her 20's", color="r",
           linestyle="dashed")
plt.ylim([0,5])
plt.xlabel("Song Periods")
plt.ylabel("Recalled Songs Ratings")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.plot(df_age_3.index, df_age_3["Recalled Songs"], color="b")
plt.axvline(df_age_3["Parent 1 Birthdate"].mean()+20,label="Parent 1 at his/her 20's",color="r",linestyle="dashed")
plt.axvline(df_age_3["Parent 2 Birthdate"].mean()+20,label="Parent 2 at his/her 20's", color="r",
           linestyle="dashed")
plt.ylim([0,5])
plt.xlabel("Song Periods")
plt.ylabel("Recalled Songs Ratings")
plt.legend()
plt.show()
```
