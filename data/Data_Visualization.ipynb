# 3. Vividness and Evoked Memories

```python
x_axis = ["1970-1974",
         "1975-1979",
         "1980-1984",
         "1985-1989",
         "1990-1994",
         "1995-1999",
         "2000-2004",
         "2005-2009"]
```

```python
df_song_recall
```

```python
df_song_recall["Evoke Memory"] = df_song_recall["Evoke Memory"]*100
```

```python
df_vividness = pd.read_csv("SongPeriods_DependentVar_MEAN_vividness.csv")
```

```python
df_vividness.set_index("Unnamed: 0", inplace=True)
```

```python
df_vividness["Vividness"] = df_vividness["Vividness"] *100
```

## Correct the y-axis and x-axis namings?

```python
fig, ax = plt.subplots(figsize=(15,10))
ind = np.arange(8)
ax = fig.add_subplot(111)
ax.bar(ind, df_song_recall["Evoke Memory"], color="c", label="Evoked Memories", width=0.27)
ax.bar(ind+0.3, df_vividness["Vividness"], color="r", label="Vividness", width=0.27)
ax.plot(ind,df_song_recall["Evoke Memory"], color="b")
ax.set(xlabel=None)
ax.set_xticklabels(list(range(1970,2010,5)))

plt.ylabel("Percentage of Participants (%)")
plt.xlabel("5 Year Period")
plt.legend()
plt.title("Percentage of Participants That Recalled Evoked Memories and Their Vividness")
plt.show()
```
