# Data Visualization of the Results

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
```

```python
participant_stat = pd.read_csv("Participant_Descriptive_Stat.csv")
```

```python
participant_stat.set_index("Unnamed: 0",inplace=True)
```

## 1. Song Periods vs. Recalled Songs

```python
df_song_recall = pd.read_csv("SongPeriods_DependentVar_mean.csv")
```

```python
df_song_recall_std = pd.read_csv("SongPeriods_DependentVar_std.csv")
```

```python
df_song_recall.set_index("Unnamed: 0", inplace=True)
```

```python
df_song_recall.head(2)
```

### Merge Parent's Means

```python
parents_mean = (participant_stat.loc["mean","Parent 1 Birthdate"] + participant_stat.loc["mean","Parent 2 Birthdate"]) /2
```

```python
font = {"size":20}
matplotlib.rc('font',**font)
plt.figure(figsize=(20,10))
plt.plot(df_song_recall.index,df_song_recall["Recalled Songs"])
plt.axvline(participant_stat.loc["mean","Participant Birthdate"], linestyle="dashed",c="m", label="Participants Born")
plt.axvline(parents_mean+20, linestyle="dashed",c="y", label="Parents Age 20 Years")
plt.ylim([-0.1,4])
plt.xlim([1970,2005])
plt.xlabel("5 Year Period")
plt.ylabel("Ratings")
plt.title("Recalled Songs")
plt.legend()
plt.show()
```

## 2. Song Periods vs. Listening Context of Memories

```python
matplotlib.rc('font',**font)
plt.figure(figsize=(20,10))
plt.plot(df_song_recall.index,df_song_recall["Childhood"], label="During Childhood")
plt.plot(df_song_recall.index,df_song_recall["Recent"], label="Recently")
plt.plot(df_song_recall.index,df_song_recall["Parents"], label="With Parents")
plt.plot(df_song_recall.index,df_song_recall["Other People"], label="With Other People")
plt.plot(df_song_recall.index,df_song_recall["Alone"], label="Alone")
plt.xlim([1970,2005])

plt.axvline(participant_stat.loc["mean","Participant Birthdate"], linestyle="dashed",c="m", label="Participants Born")
plt.axvline(parents_mean+20, linestyle="dotted",c="c", label="Parents Age 20 Years")
plt.ylim([-0.2,4])
plt.xlabel("5 Year Period")
plt.ylabel("Ratings")
plt.legend(bbox_to_anchor=(1,1), loc="upper left")
plt.title("Listening Context of Memories")
plt.show()
```

# 3. Vividness and Evoked Memories

```python
df_vividness = pd.read_csv("SongPeriods_DependentVar_MEAN_vividness.csv")
df_vividness.set_index("Unnamed: 0", inplace=True)
```

```python
df_song_recall["Evoke Memory"] = df_song_recall["Evoke Memory"]*100
df_vividness["Vividness"] = df_vividness["Vividness"] *100
```

## Correct the y-axis and x-axis namings?

```python
font = {"size":15}
matplotlib.rc('font',**font)
fig, ax = plt.subplots(figsize=(15,8))
ind = np.arange(8)
width = 0.27
ax.bar(ind, df_song_recall["Evoke Memory"], color="c", label="Evoked Memories", width=width)
ax.bar(ind+0.3, df_vividness["Vividness"], color="r", label="Vividness", width=width)
ax.plot(ind,df_song_recall["Evoke Memory"], color="b")
ax.set_xticks(ind + width / 2)
ax.set_xticklabels(list(range(1970,2010,5)))

plt.ylim([0,100])
plt.ylabel("Percentage of Participants (%)")
plt.xlabel("5 Year Period")
plt.legend()
plt.title("Percentage of Participants That Recalled Evoked Memories and Their Vividness")
plt.show()
```

# 4. Age Gap

```python
df_age_1 = pd.read_csv("AgeGap_1.csv")
df_age_2 = pd.read_csv("AgeGap_2.csv")
```

```python
df_age_1.head(3)
```

```python
df_age_2.head(2)
```

```python
df_age_1.set_index("Unnamed: 0", inplace=True)
df_age_2.set_index("Unnamed: 0", inplace=True)
```

```python
plt.figure(figsize=(15,5))
plt.plot(df_age_1.index, df_age_1["Recalled Songs"], color="c")
plt.axvline(df_age_1["Parent 1 Birthdate"].mean()+20,label="Parent 1 at his/her 20's",color="c",linestyle="dashed")
plt.axvline(df_age_1["Parent 2 Birthdate"].mean()+20,label="Parent 2 at his/her 20's", color="c",
           linestyle="dashed")
plt.xlabel("5 Year Period")
plt.ylabel("Ratings")
plt.title("Age Gap: 13")
plt.ylim([0,5])
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15,5))

plt.plot(df_age_2.index, df_age_2["Recalled Songs"], color="c")
plt.axvline(df_age_2["Parent 1 Birthdate"].mean()+20,label="Parent 1 at his/her 20's",color="r",linestyle="dashed")
plt.axvline(df_age_2["Parent 2 Birthdate"].mean()+20,label="Parent 2 at his/her 20's", color="r",
           linestyle="dashed")
plt.ylim([0,5])
plt.xlabel("5 Year Period")
plt.ylabel("Ratings")
plt.legend()
plt.title("Age Gap: 9")
plt.show()
```
